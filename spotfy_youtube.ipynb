```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sadece Spotify ve YouTube dinlenme verilerini içeren bir DataFrame oluşturalım
df_comparison = df[['Track', 'Stream', 'Views']].copy()

# İlk 20 şarkıyı alarak görselleştirme (isteğe göre artırılabilir)
df_comparison = df_comparison.head(20)

# Veriyi 'Track' üzerinden uzun formata dönüştürelim
df_melted = pd.melt(df_comparison, id_vars='Track', value_vars=['Stream', 'Views'], 
                    var_name='Platform', value_name='Plays')

# Görselleştirme
plt.figure(figsize=(14,7))
sns.barplot(x='Track', y='Plays', hue='Platform', data=df_melted, palette="viridis")
plt.xlabel('Track Name', fontsize=12)
plt.ylabel('Plays (Streams/Views)', fontsize=12)
plt.title('Spotify Streams vs YouTube Views for Tracks', fontsize=16)
plt.xticks(rotation=90)  # Şarkı isimlerini döndürme
plt.tight_layout()
plt.show()


```

```python
total_spotify = df['Stream'].sum()
total_youtube = df['Views'].sum()

# Donut Chart için veriyi hazırlayalım
data = pd.Series([total_spotify, total_youtube], index=['Spotify', 'YouTube'])

# Donut Chart görselleştirmesi
plt.figure(figsize=(8,8))
plt.pie(data, labels=data.index, autopct='%1.1f%%', startangle=140, colors=['#1DB954', '#FF0000'], wedgeprops=dict(width=0.3))
plt.title('Total Streaming Counts: Spotify vs YouTube', fontsize=16)
plt.show()
```

```python
fig, axs = plt.subplots(nrows=4, ncols=3, figsize=(15, 12))
fig.subplots_adjust(hspace=0.5)

# List of numerical features
numerical_features = ['Speechiness', 'Acousticness', 'Instrumentalness', 
                      'Liveness', 'Valence', 'Tempo', 'Duration_ms', 
                      'Views', 'Likes', 'Comments', 'Stream']

# Plot box plots for each numerical feature
for i, feature in enumerate(numerical_features):
    row = i // 3
    col = i % 3
    sns.boxplot(data=df[feature], ax=axs[row, col], color='skyblue')
    axs[row, col].set_title(feature)
    axs[row, col].set_xlabel('')
    axs[row, col].set_ylabel('')
    axs[row, col].tick_params(axis='x', labelrotation=45)

# Remove empty subplots
for i in range(len(numerical_features), len(axs.flatten())):
    fig.delaxes(axs.flatten()[i])

# Show the plot
plt.tight_layout()
plt.show()
```

```python
sum_album_type = df['Album_type'].value_counts()
#Adds color to pie chart
colors = sns.color_palette()[0:3]
#Plots pie chart
plt.pie(sum_album_type.values, labels = sum_album_type.index, colors = colors, autopct = '%.0f%%', shadow = True)
plt.title('Percentage of Count Values for Each Album Type')
#Edit font size
plt.rcParams['font.size'] = 12
```

```python

#Group in descending order of total views by artist.
artist_views = df.groupby('Artist', as_index = False).sum().sort_values(by = 'Views', ascending = False)
fig, ax = plt.subplots(figsize = (8,10))
#Create barplot of top 10 artists by views.
sns.barplot(x = artist_views['Views'][:10], y = artist_views['Artist'][:10])
plt.xticks(rotation = 90)
ax.set_title('Top 10 Artists by Views')
ax.set_xlabel('Total Views (Billions)')
```

```python
song_views = df.groupby('Track', as_index = False).sum().sort_values(by = 'Views', ascending = False)
fig, ax = plt.subplots(figsize = (8,10))
#Create barplot of top 10 artists by views.
sns.barplot(x = song_views['Views'][:10], y = song_views['Track'][:10])
plt.xticks(rotation = 90)
ax.set_title('Top 10 Songs by Views')
ax.set_xlabel('Total Views (Billions)')
```

```python
artist_likes = df.groupby('Artist', as_index = False).sum().sort_values(by = 'Likes', ascending = False)
fig, ax = plt.subplots(figsize = (8,10))
#Create barplot of top 10 artists by likes.
sns.barplot(x = artist_likes['Likes'][:10], y = artist_likes['Artist'][:10])
plt.xticks(rotation = 90)
ax.set_title('Top 10 Artists by Likes')
ax.set_xlabel('Total Likes (Billions)')
```

```python
#Group in descending order of total views by artist and track.
puth_df = df.groupby(['Artist', 'Track'], as_index = False).sum()
#Sort values by the total views with Charlie Puth as the artist. 
puth_views = puth_df[puth_df['Artist'] == 'Charlie Puth'].sort_values(by = 'Views', ascending = False)
#Create barplot of top 5 Charlie Puth songs by the total amount of views on YouTube.
fig, ax = plt.subplots(figsize = (8,10))
sns.barplot(x = puth_views['Views'][:5], y = puth_views['Track'][:5])
plt.xticks(rotation = 90)
ax.set_title('Top 5 Charlie Puth Songs by Views')
ax.set_xlabel('Total Views (Billions)')
```
